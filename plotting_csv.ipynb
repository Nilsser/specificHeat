**Plotten der CSV Dateiena**

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import cm
import numpy as np
import csv
import pandas as pd
import matplotlib.ticker as ticker
from matplotlib.ticker import FuncFormatter
from matplotlib.ticker import MaxNLocator
from matplotlib.ticker import *
from matplotlib import gridspec
```

*Read in the data from our three .csv files*

```python
aluminium = pd.read_csv('data/Copper_Linsi_Serck.csv')
print(aluminium.head())
#aluminium=aluminium.iloc[::100, ]
time = aluminium["Time"]
channelA = aluminium["Channel A"]
channelB = aluminium["Channel B"]
```

```python
step =1 #only plotting every n-th point to avoid overplotting

x_ticker_count = 5
y_ticker_count = 4
```

```python

fig = plt.figure(figsize=(15, 10)) 
gs = gridspec.GridSpec(2, 1, height_ratios=[3, 1]) 
ax0 = plt.subplot(gs[0])
ax1 = plt.subplot(gs[1])

ax0.plot(time[::step], channelA[::step],'o',alpha=0.2,color='b',)
ax0.set_title("PT100 [mV]")
#ax[0].legend()

ax0.xaxis.set_major_locator(MaxNLocator(x_ticker_count))  # Reduce number of x-axis ticks
ax0.yaxis.set_major_locator(MaxNLocator(y_ticker_count))  # Limit to 5 major y-axis ticks

#ax0.xaxis.set_major_formatter(FuncFormatter(seconds_to_minutes))
#ax0.yaxis.set_major_formatter(FuncFormatter(cut))

ax1.plot(time[::step], channelB[::step],'-',color='r',linewidth=0.7)
ax1.set_title("Heating Voltage [V]")
#ax[1].legend()

ax1.xaxis.set_major_locator(MaxNLocator(x_ticker_count))  # Reduce number of x-axis ticks
ax1.yaxis.set_major_locator(MaxNLocator(y_ticker_count))  # Limit to 5 major y-axis ticks
#ax1.xaxis.set_major_formatter(FuncFormatter(seconds_to_minutes))




plt.show()
```
