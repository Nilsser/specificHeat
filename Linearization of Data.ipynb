# Initialize

## Load libraries

```python
import matplotlib
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig_width = 7
fig_height = 6 
fs = 18
ls = 16
```

# Data

## Generate data

```python
# Number of points
n = 100

# Generate random measurement errors
std = 0.01
e = np.random.normal(0,std,n)

x = np.linspace(0,10,n)

# Exponential decay
a = 1
b = 0.1
y_1 = np.exp(-a*x) + e
y_2 = np.exp(-b*x) + e
# Biexponential decay
y_3 = 0.5*(y_1+y_2)
```

## Plot data

```python
plt.figure(
    num=1,
    figsize=(fig_width, fig_height),
    facecolor='white')
plt.xlabel('x', fontsize=fs)
plt.ylabel('y', fontsize=fs)
plt.plot(x,y_1,'bo', label='Exponential')
plt.plot(x,y_2,'ro', label='Exponential')
plt.plot(x,y_3,'go', label='Biexponential')
plt.tick_params(axis='both', which='major', labelsize=ls)
plt.legend()
plt.show()
```

```python
plt.figure(
    num=1,
    figsize=(fig_width, fig_height),
    facecolor='white')
plt.xlabel('x', fontsize=fs)
plt.ylabel('y', fontsize=fs)
plt.xscale('linear')
plt.yscale('log')
plt.plot(x,y_1,'bo', label='Exponential')
plt.plot(x,y_2,'ro', label='Exponential')
plt.plot(x,y_3,'go', label='Biexponential')
plt.tick_params(axis='both', which='major', labelsize=ls)
plt.legend()
plt.show()
```
